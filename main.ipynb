```python
import lda
import lda.datasets
```

```python
X = lda.datasets.load_reuters()
vocab = lda.datasets.load_reuters_vocab()
titles = lda.datasets.load_reuters_titles()
```

```python
def matrix_to_documents(doc_term_matrix, vocabulary):
    documents = []
    for row in doc_term_matrix:
        doc = []
        for term_index, term_count in enumerate(row):
            for _ in range(term_count):  # Add each term the number of times it appears
                doc.append(vocabulary[term_index])
        documents.append(doc)
    return documents

# Convert the document-term matrix to lists of words
converted_documents = matrix_to_documents(X, vocab)

# Now that we have the documents in the right format, we can apply the CollapsedVB class
K = 10  # You can choose the number of topics
cvb = CollapsedVB(converted_documents, K)
cvb.run(100)

theta_dk = cvb.get_topic_distributions()
phi_kv = cvb.get_word_distributions()

print("Topic distributions per document:", theta_dk)
print("Word distributions per topic:", phi_kv)
```
